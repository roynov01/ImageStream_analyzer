# Image Stream Analysis Example

This notebook demonstrates image loading, visualization, and basic analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../src')

from visualization import display_image, InteractiveImageViewer
from image_handler import load_ome_tiff, save_image
```

## Create a sample image

```python
# Create a sample image for demonstration
x = np.linspace(-10, 10, 512)
y = np.linspace(-10, 10, 512)
X, Y = np.meshgrid(x, y)
sample_image = np.sin(np.sqrt(X**2 + Y**2) / 2) * 255
sample_image = (sample_image - sample_image.min()) / (sample_image.max() - sample_image.min()) * 255

print(f"Sample image shape: {sample_image.shape}")
print(f"Data range: {sample_image.min():.2f} - {sample_image.max():.2f}")
```

## Display the image

```python
display_image(sample_image, title="Sample Image", cmap="viridis")
```

## Interactive image viewer with click detection

```python
# Create and display interactive viewer
viewer = InteractiveImageViewer(sample_image, title="Click on the image to detect positions")
# Note: Call viewer.show() to display - you can click on the image and the coordinates will be printed
```
